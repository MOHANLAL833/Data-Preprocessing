```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy="median")
imp.fit(data[quan])
quan_data=imp.transform(data[quan])
```

```python
quan_data=pd.DataFrame(quan_data,columns=quan)
quan_data
```

```python
#categorical
import numpy as np
from sklearn.impute import SimpleImputer
imp= SimpleImputer(strategy='most_frequent')
imp.fit(data[qual])#we take quan becoz the dataset having character also so it shows error,
#considering that we take quan for values only
qual_data=imp.transform(data[qual])
```

```python
qual_data=pd.DataFrame(qual_data,columns=qual)
qual_data
```

```python
merge=[quan_data,qual_data]
merge
```

```python
preprocessed=pd.concat(merge,axis=1)
preprocessed
```
